loading the model

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
from gtts import gTTS
from IPython.display import Audio, display

# Load Crisbee's brain (change model name here if needed)
model_name = "distilgpt2"  # light and fast; try 'microsoft/phi-2' for smarter responses (slower though)
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

```

Crisbee Response Function

```python
def crisbee_think(prompt, max_length=150):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=max_length, do_sample=True, top_p=0.9, temperature=0.7)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response[len(prompt):].strip()
```

Crisbee Speaks - Text to Speech

```python
def crisbee_speak(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save("crisbee_response.mp3")
    display(Audio("crisbee_response.mp3", autoplay=True))
```

Chat with Crisbee

```python
# Interact with Crisbee
user_input = input("You : ")
crisbee_response = crisbee_think(f"You: {user_input}\nCrisbee: ")

print(f"Crisbee : {crisbee_response}")
crisbee_speak(crisbee_response)
```
